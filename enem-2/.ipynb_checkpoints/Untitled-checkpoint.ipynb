## Verificação  dos dados de Treinos e de Teste

Inicialmente, é interessante que vejamos se os datasets de treino e teste 'batem', isto é, se ambos apresentam as mesmas colunas ou que um seja subconjunto da outra.

```python
train = pd.read_csv('train.csv')
test =  pd.read_csv('test.csv')
print(set(test.columns).issubset(set(train.columns)))
print(len(train.columns), len(test.columns))
```

Os valores apresentam uma importante informação: os dados de teste são bem mais limitados que os de treino: os dados de teste tem apenas 47 colunas comparadas as 167 dos dados de treino. Portanto, podemos admitir que embora se possa eventualmente retirar informações das colunas restantes, não poderemos utilizar esse conhecimento nos dados de teste. Nessas condições, é conveniete **limitarmos o conjunto de treino às mesmas colunas observadas nos dados de teste.**

Obviamente, para podermos treinar é importante termos informações da variável que queremos prever; sendo assim, incluimos a variável  `NU_NOTA_MT` no dataset de treino também.

```python
features_preliminares = list(test.columns)+['NU_NOTA_MT']
enem = train[features_preliminares]
enem.shape
```

Muito bom! Isso já nos reduz a cerca de 25% das variáveis que tínhamos antes. Vamos agora observar que variáveis são essas e quais os tipos de dados que as formam; Também é importante avaliar se existem muitos **missing values**, os *valores faltantes* que muito atrapalham as predições.

```python
enem.columns
```

```python
def aux_df(df):

    aux = pd.DataFrame({#'coluna': enem.columns,
                        'tipos': enem.dtypes,
                        'NA#': enem.isna().sum(),
                        'NA%': enem.isna().sum()/enem.shape[0],
                        'qtdCat': enem.nunique()
    })

    
    return aux.sort_values(by=['NA%','tipos'],ascending=False)

aux_train = aux_df(enem)

aux_train
```

```python
def nan_plot(df_aux):
    plt.figure(figsize=(20,15))
    return sns.barplot(x ='NA%', y =df_aux.index, data=df_aux)

nan_plot(aux_train)    
```

Grande parte das variáveis é do tipo categórica. Isso pode ser visto  no Dicionário dos Dados (link aqui). nos nomes das variáveis, `IN`,`CO`,`NU` e `SG` significam respectivamente *Indicador, Código, Número* e *Sigla*; Não consegui identificar exatamente o exado significado da sigla `TP`. Além disso, algumas das linhas apresentam valores faltantes. Vamos examinar as quantidades deles e resolver como tratar esses dados

```python
aux_teste = aux_df(test)

aux_teste
```

As colunas *Q027, TP_ENSINO, TP_DEPENDENCIA_ADM_ESC* apresentam mais da metade dos seus valores como faltantes. Nesse caso, é difícil acrescentar completá-las para poder extrair algo. Adicionalmente', as variáveis de notas apresentam quase o mesmo valor de notas faltantes entre si, cerca de 26%. Isso inclui nossa variável alvo.

Antes de fazer qualquer coisa, vamos analisar também o dataset de testes

```python
nan_plot(aux_teste)
```

Pois bem. Agora, vamos analisar as variáveis numéricas que os datasets apresentam, buscando entender mais alguma coisa sobre elas. É util nesse caso fazer uma separação entre as variáveis categóricas e núméricas, para lidar separadamente com as variáveis conforme for o caso.

```python
cat_prefix = ['CO','IN','TP','SG','Q0']
cats = []
nums = []
#print(len(features_preliminares))
for feature in features_preliminares:
    if feature[:2] in cat_prefix:
        cats.append(feature)
    else:
        nums.append(feature)
nums.remove('NU_INSCRICAO')
nums
```

```python
def dist(df,s):
    fig = plt.figure()
    sns.boxplot(df[s])
    return fig
```

```python
enem[nums].plot(kind='box', y=nums,figsize=(45,15), fontsize=24)
```
